```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("prosodic_features.csv")
labels = pd.read_csv("turker_scores_full_interview.csv")
transcripts = pd.read_csv("interview_transcripts_by_turkers.csv")
```

```python
df["participant&question"].unique()
```

```python
df["Total"].unique()
```

```python
for col in labels.columns:
    if(col!="Participant" and col !="EyeContact" and col!="Smiled"):
        df[col]= 0
df["question_number"] = None
df.columns
```

```python
flag = "ok"
df["Overall"] = None
for ind in df.index:
    df["participant&question"][ind] = df["participant&question"][ind].lower()
    part = df["participant&question"][ind][:df["participant&question"][ind].find("q")]
    df["question_number"] =  df["participant&question"][ind][df["participant&question"][ind].find("q"):]
    df["participant&question"][ind] = part
    for i in labels.index:
        if(flag==labels["Participant"][i]): continue
        if(part==labels["Participant"][i]):
            flag = labels["Participant"][i]
            for col in df[['Worker', 'Colleague',
                           'Engaged', 'Excited', 'SpeakingRate','NoFillers', 'Friendly', 'Paused', 'EngagingTone', 'StructuredAnswers',
                           'Calm', 'NotStressed', 'Focused', 'Authentic', 'NotAwkward','Overall','RecommendHiring', 'Total']]:
                df[col][ind] = labels[col][i]
            break
```

```python
for ind in df.index:
    for col in df.columns:
        if(df[col][ind]==None or (df[col][ind]==0 and (col=="Total" or col=="RecommendHiring" or col=="Overall"))):
            df[col][ind] = df[col][ind-1]
```

```python
df.to_csv("full_NLP_df.csv")
df
```

```python
cols =  df.drop(["participant&question","question_number"], axis=1).copy().columns
mean_df = pd.DataFrame(index=labels.Participant.unique(), columns = cols)
for col in mean_df.columns:
    mean_df[col]=0.0

mean_df
```

```python
for i, part in enumerate(mean_df.index):
    counter = 0 
    for ind in df.index:
        if(part==df["participant&question"][ind]):
            counter += 1
            for col in mean_df.columns:
                if(col=="Total" or col=="RecommendHiring" or col=="Overall"): mean_df[col][i] = df[col][ind]
                else:
                    if(type(df[col][ind])!=str): mean_df[col][i] += df[col][ind]
    if(counter!=0):
        for col in mean_df.columns:
            if(col!="Total" and col!="RecommendHiring" and col!="Overall"): mean_df[col][i] = mean_df[col][i]/counter
```
